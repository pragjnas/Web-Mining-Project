### Calc the percentage_score
##### percentage_score = num of reviews per prod/ Total # of reviews for all products

```python
grouped_sum = sorted_data['score_counts'].sum()
print (grouped_sum)
sorted_data['percentage']  = sorted_data['score_counts'].div(grouped_sum)*100
product_mean_count_percent = sorted_data.sort_values(['score_counts', 'ProductId'], ascending = [0,1])
print(product_mean_count_percent)
```

### Merging the main data set with the product_mean_count_percent data frame on Product Id

```python
merged_dataset = pd.merge(amazon_data_set, product_mean_count_percent, on='ProductId')
len(merged_dataset)
merged_dataset.head()
```

```python
len(merged_dataset)
```

### Mimimizing the data set to reduce the noise i.e. removing the products that have less than 5% of the total ratings

```python
minimized_data_set = merged_dataset[(merged_dataset.Mean_Score>1) & (merged_dataset.percentage>0.05)]
len(minimized_data_set)
```

```python
# Exporting the minimized_data_set to csv
export_csv = amazon_data_set.to_csv (r'C:\Users\jyost\OneDrive\Desktop\Overall Info\Imp\Fall 2019\Project\Amazon Fine Food\minimized_data_set.csv', index = None, header=True) #Don't forget to add '.csv' at the end of the path
```

### Pivoting the data --> Columns: Prod ID; Rows: User ID that shows the rating given by a user to a specific product

```python
# product_matrix = top_products.pivot_table(index='UserId', columns='ProductId', values='Score', aggfunc=[np.mean])
product_matrix = minimized_data_set.pivot_table(index='UserId', columns='ProductId', values='Score')

product_matrix
```

### Taking the B007JFMH8M for correlation as it is having the highest number of reviews and a good mean score 

```python

B007JFMH8M_ratings  = product_matrix['B007JFMH8M']
# B007JFMH8M_ratings = pd.DataFrame(B007JFMH8M_ratings, columns=['B007JFMH8M'])

B007JFMH8M_ratings
```

### Correlationfor the B007JFMH8M vs other products

```python
product_like_B007JFMH8M = product_matrix.corrwith(B007JFMH8M_ratings)
product_like_B007JFMH8M
```

### Sorting the correlation values of the product B007JFMH8M in the descending order

```python
corr_B007JFMH8M = pd.DataFrame(product_like_B007JFMH8M, columns=['Correlation'])
corr_B007JFMH8M.dropna(inplace=True)
corr_B007JFMH8M.sort_values('Correlation', ascending=False).head(10)
```
